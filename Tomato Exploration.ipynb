```python
import pickle
import numpy as np

with open("clean_dictionary.pkl", "rb") as file:
    clean_dicti = pickle.load(file)
    
clean_dicti = np.array(clean_dicti)
```

```python
with open("clean_snippets.pkl", "rb") as file:
    clean_snippets = pickle.load(file)
```

```python
clean_dicti_list = []

for i in range(clean_dicti.shape[0]):
    clean_dicti_list.append(list(clean_dicti[i]))
```

```python
from collections import Counter
import itertools

word_count_dic = Counter(itertools.chain.from_iterable(clean_dicti_list))
vocab_freq_dic = word_count_dic.most_common()[1:]
vocab_dic = [k for k,v in word_count_dic.most_common()[1:]]
word_freq_dic = [v for k,v in word_count_dic.most_common()[1:]]

word_count_snip = Counter(itertools.chain.from_iterable(clean_snippets))
vocab_freq_snip = word_count_snip.most_common()[1:]
vocab_snip = [k for k,v in word_count_snip.most_common()[1:]]
word_freq_snip = [v for k,v in word_count_snip.most_common()[1:]]
```

```python
### Word frequency line plot
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 2, figsize=(14, 7))


ax[0].plot(np.linspace(0, len(word_freq_dic), len(word_freq_dic)), np.array(word_freq_dic))

ax[0].set_xscale("linear")
ax[0].set_yscale("log")

ax[0].set_xlabel("Each word", fontdict={'fontsize': 18, 'fontweight': 'roman'})
ax[0].set_ylabel("Number of occurences (log scale)", fontdict={'fontsize': 18, 'fontweight': 'roman'})
ax[0].set_title("Word frequency distribution - Dictionary   ", fontdict={'fontsize': 22, 'fontweight': 'roman'})


ax[1].plot(np.linspace(0, len(word_freq_snip), len(word_freq_snip)), np.array(word_freq_snip))

ax[1].set_xscale("linear")
ax[1].set_yscale("log")

ax[1].set_xlabel("Each word", fontdict={'fontsize': 18, 'fontweight': 'roman'})
ax[1].set_ylabel("Number of occurences (log scale)", fontdict={'fontsize': 18, 'fontweight': 'roman'})
ax[1].set_title("   Word frequency distribution  - Original Snippets", fontdict={'fontsize': 22, 'fontweight': 'roman'})

plt.tight_layout()
plt.savefig("word freq dist.png")
plt.show()
```
